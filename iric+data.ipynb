```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv("iris-data.csv")
```

```python
data
```

```python
sedata=data[data.iris_class == 'Iris-setosa']
vedata=data[data.iris_class == 'Iris-versicolor']
vidata=data[data.iris_class == 'Iris-virginica']
```

```python
selen=np.array(sedata['petal_length_cm'])
vilen=np.array(vidata['petal_length_cm'])
velen=np.array(vedata['petal_length_cm'])
```

```python
def el(data):
    n=len(data)
    
    x=np.sort(data)
    
    y=np.arange (1,n+1)/n
    
    return x ,y
```

```python
semean=np.mean(selen)
vimean=np.mean(vilen)
vemean=np.mean(velen)

sestd=np.std(selen)
vistd=np.std(vilen)
vestd=np.std(velen)

semedian=np.median(selen)
vimedian=np.median(vilen)
vemedian=np.median(velen)

x,y=el(selen)
print(x)
print(y)
a,b=el(vilen)
print(a)
print(b)
z,w=el(velen)
print(z)
print(w)
```

```python
allmean=[semean,vimean,vemean]
all_mean=np.mean(allmean)

allmedian=[semedian,vimedian,vemedian]
all_median=np.median(allmedian)

allstd=[sestd,vistd,vestd]
all_std=np.std(allstd)
```

```python
# drow the chart
_=plt.plot(x,y,marker='.',color='red' , linestyle='none')
_=plt.plot(a,b,marker='*',color='green' ,  linestyle='none')
_=plt.plot(z,w,marker='.',color='black' ,  linestyle='none')

plt.margins(0.2)
_=plt.plot((all_mean,all_mean),(0.1,0.9),marker='.',color='brown')
_=plt.plot((all_median,all_median ),(0.1,0.9), marker='_', color = 'gray')
_=plt.plot((0,6),(all_std, all_std), marker='_', color = 'blue')

_ = plt.legend(('seto','virg','vers'), loc='upper left')
_ = plt.xlabel('higth (cm)')
_ = plt.ylabel('ECDF')

plt.show()

```
