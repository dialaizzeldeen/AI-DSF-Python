```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
data =pd.read_csv('iris-data.csv')
```

```python
data
```

```python
setodata =data[data.iris_class=='Iris-setosa']
```

```python
versdata= data[data.iris_class == 'Iris-versicolor']
```

```python
virdata=data[data.iris_class=='Iris-virginica']
```

```python
setlen=np.array(setodata['petal_length_cm'])
setwid=np.array(setodata['petal_width_cm'])
verslen=np.array(versdata['petal_length_cm'])
verswid=np.array(versdata['petal_width_cm'])
virlen=np.array(virdata['petal_length_cm'])
virwid=np.array(virdata['petal_width_cm'])
```

```python
# Make a scatter plot
_ = plt.plot(verslen, verswid,
             marker='.', linestyle='none')

# Set margins
plt.margins(0.02)

# Label the axes
_ = plt.xlabel('petal length (cm)')
_ = plt.ylabel('petal width (cm)')

# Show the result
plt.show()
```

```python
# Make a scatter plot
_ = plt.plot(virlen, virwid,
             marker='.', linestyle='none')

# Set margins
plt.margins(0.02)

# Label the axes
_ = plt.xlabel('petal length (cm)')
_ = plt.ylabel('petal width (cm)')

# Show the result
plt.show()
```

```python
# Make a scatter plot
_ = plt.plot(setlen, setwid,
             marker='.', linestyle='none')

# Set margins
plt.margins(0.02)

# Label the axes
_ = plt.xlabel('petal length (cm)')
_ = plt.ylabel('petal width (cm)')

# Show the result
plt.show()
```

```python
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""

    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n

    return x, y
```

```python


verpetal_mean = np.mean(verslen)
verpetal_median = np.median(verslen)
verpetal_std = np.std(verslen)

virpetal_mean = np.mean(virlen)
virpetal_median = np.median(virlen)
virpetal_std = np.std(virlen)

setpetal_mean = np.mean(setlen)
setpetal_median = np.median(setlen)
setpetal_std = np.std(setlen)


x, y = ecdf(verslen)
w, z = ecdf(virlen)
a, b = ecdf(setlen)
_=plt.plot(x, y, marker='.',  linestyle='none' , color='blue')
_=plt.plot(w, z, marker='*' , linestyle='none', color='brown')
_=plt.plot(a, b, marker='.',  linestyle='none', color='orange')

plt.margins(0.02)
_=plt.plot((verpetal_mean,verpetal_mean),(0.1,0.9), marker='.', color = 'red')
_=plt.plot((verpetal_median,verpetal_median),(0.1,0.9), marker='_', color = 'black')
_=plt.plot((0,6),(verpetal_std, verpetal_std), marker='_', color = 'green')

_=plt.plot((virpetal_mean,virpetal_mean),(0.1,0.9), marker='.', color = 'yellow')
_=plt.plot((virpetal_median,virpetal_median),(0.1,0.9), marker='_', color = 'pink')
_=plt.plot((0,6),(virpetal_std, virpetal_std), marker='_', color = 'blue')

_=plt.plot((setpetal_mean,setpetal_mean),(0.1,0.9), marker='.', color = 'brown')
_=plt.plot((setpetal_median,setpetal_median),(0.1,0.9), marker='_', color = 'gray')
_=plt.plot((0,6),(setpetal_std, setpetal_std), marker='_', color = 'orange')




_ = plt.legend(('vers', 'vir', 'set'), loc='upper left')
_=plt.xlabel('length')
_=plt.ylabel('ECDF')
plt.show()





```
